vetores não ordenados:

```python
class VetorNaoOrdenado:
  def __init__(self, capacidade):
    self.capacidade = capacidade
    self.ultima_posicao = -1
    self.valores = [''] * capacidade

  def inserir (self, valor):
    if self.ultima_posicao == self.capacidade - 1:
      print("vetor cheio")
    else:
      self.ultima_posicao += 1 #incremento
      self.valores[self.ultima_posicao] = valor

  def imprimir(self):
    if self.ultima_posicao == -1:
      print("vetor vazio")
    else:
      for i in range(self.ultima_posicao + 1):
        print("Posição {} | Valor {}".format(i, self.valores[i]))

  def pesquisar(self, valor):
    for i in range(self.ultima_posicao + 1):
      if self.valores[i] == valor:
        return i
    return -1

  def excluir(self, valor):
    posicao = self.pesquisar(valor)
    if posicao == -1:
      return -1
    else:
      for i in range(posicao, self.ultima_posicao): #o range vai da posicao do valor informado até a ultima posicao
        self.valores[i] = self.valores[i+1] #pega o numero que ta na posicao ao lado e coloca na do que foi excluido e assim por diante
        self.ultima_posicao -= 1

meu_vetor = VetorNaoOrdenado (4)
meu_vetor.imprimir()
meu_vetor.inserir(2)
meu_vetor.inserir(17)
meu_vetor.inserir(7)
meu_vetor.inserir(25)
meu_vetor.inserir(66) #print vetor cheio
meu_vetor.imprimir()
print(meu_vetor.pesquisar(17)) #posicao 1
print(meu_vetor.pesquisar(35)) #-1 pq nao existe
print(meu_vetor.pesquisar(25)) #posicao 3
meu_vetor.excluir(17)
meu_vetor.imprimir()
print(meu_vetor.pesquisar(7))
meu_vetor.imprimir()


```
